```python
import pandas as pd
from pymongo import MongoClient
import pymongo
import os
from dotenv import load_dotenv
load_dotenv()
import sys
sys.path.append('../src')
import apilast as apl
import sqltools as sqt
import random
import requests
```

```python
client = MongoClient("localhost:27017")
lastusers = client.get_database("lastusers")
```

```python
lastusers.list_collection_names()
```

```python
pablo = lastusers.pablicic
```

```python
pablo
```

```python
pablo_base = list(pablo.find())
```

```python
# sampleando = random.choices(pablo_base, k=30)
```

```python
lista = []
dic_bas = {}
for t in pablo_base:
    dic_bas = {}
    dic_bas['uts'] = t['date']['uts']
    dic_bas['fecha'] = t['date']['#text']
    dic_bas['artist'] = t['artist']['#text']
    dic_bas['artist_mbid'] = t['artist']['mbid']
    dic_bas['image'] = t['image'][2]['#text']
    dic_bas['album'] = t['album']['#text']
    dic_bas['album_mbid'] = t['album']['mbid']
    dic_bas['title'] = t['name']  
    dic_bas['track_mbid'] = t['mbid']      
    dic_bas['url'] = t['url']
    lista.append(dic_bas)
    
```

```python
lista[0:10]
```

```python
pablodf = pd.DataFrame(lista)
```
