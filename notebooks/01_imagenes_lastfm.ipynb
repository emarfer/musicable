````python
import os
import requests
import sys
sys.path.append('../src')
import sqltools as sqt

from PIL import Image
from io import BytesIO

import random
````

````python
# token lastfm
keylast = os.getenv("keylast")
````

````python
l_user = 'sinatxester' #usuario lastfm
limit = 1000 #limite de resultados por página de la request
uts_num = sqt.maxuts() + 1 #último uts para cargar datos y limitar 
````

````python
#url request
url = f'http://ws.audioscrobbler.com/2.0/?method=user.getrecenttracks&user={l_user}&limit={limit}&from={uts_num}&api_key={keylast}&format=json'
````

````python
temas = requests.get(url).json()['recenttracks']['track'] #request a la api para tener unos cuantos temas sobre los que trabajar
````

````python
temas
````

````python
tema = random.choice(temas) # genero un tema en concreto
````
