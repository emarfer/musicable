```python
url = "https://legismusic.com/es/generos-musicales/"
html = requests.get(url)
soup = BeautifulSoup(html.content,"html.parser")
```

```python
soup.find("h3").next_element.next_element.next_element
```

```python
genere_dict = {}
for s in soup.find_all("h3"):
    if s.has_attr("class") == False:
        genero = s.getText()
        genere_dict[genero] = []
        if s.find_next("ol"):
            contenedor_list_subgeneres = s.find_next("ol")
            for subgen in contenedor_list_subgeneres:
                subgenero_text = subgen.getText().strip()
                if subgenero_text != "":
                    genere_dict[genero].append(subgenero_text)
        elif s.find_next("ul"):
            contenedor_list_subgeneres = s.find_next("ul")
            for subgen in contenedor_list_subgeneres:
                subgenero_text = subgen.getText().strip()
                if subgenero_text != "":
                    genere_dict[genero].append(subgenero_text)

print (genere_dict)
genere_dict["Rock"]
```

```python
<div class="prose max-w-none dark:prose-invert sm:prose-lg prose-headings:font-normal prose-p:text-gray-600 prose-a:text-primary-500 prose-blockquote:border-l-primary-500 dark:prose-p:text-gray-300 dark:prose-a:text-primary-400 md:prose-h2:text-5xl md:prose-h3:text-4xl md:prose-h4:text-3xl md:prose-h5:text-2xl md:prose-h6:text-xl [&amp;_iframe]:aspect-video [&amp;_iframe]:w-full">  <p><img src="/_vercel/image?url=_astro%2Funlimited-downloads-03-80x80.45IOu2gz.png&amp;w=640&amp;q=100" alt="unlimited-downloads" class="float-left mb-4 mr-4" width="80" height="80" loading="lazy" decoding="async">Si oyes la palabra “música”, ¿qué canción suena en tu cabeza?</p>
<p>Probablemente sea algún género musical como pop, electrónica, reggaeton….</p>
<p>Lo que es seguro es que no pensarás en la música <em>Cruck</em> o <em>Chillwave</em>; o quizá suene en tu cabeza una canción de este estilo, pero no sepas que pertenece a un género musical con un nombre tan poco común.</p>
<p>Es muy común que identifiquemos ciertas canciones con un género musical conocido sin saber que probablemente pertenezca a un subgénero llamado diferente o incluso pertenezca a otro género.</p>
<blockquote>
<p>Esto ocurre por el simple hecho de que hay tantos estilos musicales diferentes, con sus propios subgéneros y que también varían en función de la región en la que se originaron, que es imposible conocerlos todos.</p>
</blockquote>
<p>Así pues, en este artículo hemos decidido reunir, ordenados por fecha de origen así como por región, <strong>la gran mayoría de los géneros musicales actualmente identificados.</strong></p>
<p><img src="/_vercel/image?url=_astro%2Fregistro-derechos-de-autor-espana-80x80.weUnfxQR.png&amp;w=640&amp;q=100" alt="registro derechos de autor españa" class="float-left mb-4 mr-4" width="80" height="80" loading="lazy" decoding="async">No obstante, cabe señalar que esta lista no es exhaustiva, ya que los géneros y subgéneros musicales siguen evolucionando y con el tiempo surgen otros nuevos.</p>
<p>Además, es difícil precisar las fechas exactas de origen de muchos géneros, por lo que se han indicado en su lugar periodos de tiempo aproximados.</p>
<h2 id="géneros-musicales-ordenados-por-época">Géneros musicales ordenados por época</h2>
<div class="rounded-md p-4 !my-5 bg-warning-50 text-warning-700 dark:bg-warning-950 dark:text-warning-300">  <ul>
<li><strong>Clásica</strong> (<em>entre 1750-1820</em>) – Compuesto por músicos como Mozart y Beethoven, este género se caracteriza por
melodías intrincadas y composiciones estructuradas.</li>
<li><strong>Blues</strong> (<em>finales del siglo XIX</em>) – Arraigada en las tradiciones musicales africanas, la música blues transmite emoción a través de sencillas progresiones de acordes y expresivas voces.</li>
<li><strong>Jazz</strong> (<em>principios del siglo XX</em>) – Originario de Nueva Orleans, el jazz combina el blues, el ragtime y la música de banda para crear melodías alegres e improvisadas.</li>
<li><strong>Country</strong> (<em>década de 1920</em>) – Inspirado en el folk, el blues y la música gospel, el country suele caracterizarse por letras que cuentan historias e instrumentos como el banjo y el violín.</li>
<li><strong>Swing</strong> (<em>años 30</em>) – Subgénero del jazz, el swing se caracteriza por una potente sección rítmica y arreglos de big band que incitan al baile.</li>
<li><strong>Rock ‘n’ Roll</strong> (<em>años 50</em>) – Fusionando el rhythm and blues con el country, el rock ‘n’ roll es conocido por sus ritmos enérgicos, su guitarra eléctrica y su actitud rebelde.</li>
<li><strong>Rhythm and Blues</strong> (<em>1940-1950</em>) – A menudo abreviado como R&amp;B, este género mezcla influencias del jazz, el blues y el gospel para crear melodías conmovedoras y bailables.</li>
<li><strong>Funk</strong> (<em>década de 1960</em>) – Se caracteriza por ritmos sincopados, líneas de bajo groovy y un enfoque en el ritmo. El funk surgió del soul y el jazz.</li>
<li><strong>Reggae</strong> (<em>años 60</em>) – Procedente de Jamaica, el reggae combina elementos de la música folclórica caribeña, el
calipso y el rhythm and blues, con un énfasis en los ritmos desenfrenados.</li>
<li><strong>Rock psicodélico</strong> (<em>años 60</em>) – Marcado por sonidos experimentales y letras que exploran estados alterados de conciencia. El rock psicodélico surgió durante el movimiento contracultural.</li>
<li><strong>Hard rock</strong> (<em>década de 1960</em>) – El hard rock, una forma más pesada y agresiva de la música rock, se caracteriza por guitarras distorsionadas, baterías machaconas y voces estridentes.</li>
<li><strong>Heavy Metal</strong> (<em>finales de la década de 1960-1970</em>) – Con sus orígenes en el hard rock y el blues-rock, el heavy
metal es conocido por su distorsión amplificada, largos solos de guitarra y letras agresivas.</li>
<li>**Rock progresivo **(<em>finales de la década de 1960-1970</em>) – El rock progresivo se caracteriza por la fusión del rock con otros géneros como la música clásica y el jazz, las estructuras complejas de las canciones y la instrumentación experimental.</li>
<li><strong>Disco</strong> (<em>años 70</em>) – Definida por sus ritmos contagiosos y bailables, la música disco surgió del funk, el soul y el
pop.</li>
<li><strong>Punk Rock</strong> (<em>años 70</em>) – Nacido como reacción a la música dominante, el punk rock se caracteriza por su
sonido crudo y acelerado y su actitud rebelde.</li>
<li><strong>Hip Hop</strong> (<em>finales de los 70</em>) – Con origen en el Bronx, el hip hop abarca el rap, el DJ, el breakdance y el arte del graffiti, y a menudo se centra en cuestiones sociales y políticas.</li>
<li><strong>New Wave</strong> (<em>finales de los 70-1980</em>) – Mezcla de punk rock y música pop. La nueva ola es conocida por
su sonido extravagante y experimental, que a menudo incorpora sintetizadores.</li>
<li><strong>Synthpop</strong> (<em>década de 1980</em>) – El synthpop, basado en gran medida en sintetizadores, es conocido por su sonido extravagante y experimental, y a menudo incorpora sintetizadores.</li>
<li><strong>Electronic Dance Music (EDM)</strong> (<em>finales de los 80-actualidad</em>) – Término que engloba varios géneros de música electrónica, como el house, el techno, el trance y el dubstep, caracterizados por ritmos repetitivos y sonidos sintetizados.</li>
<li><strong>House</strong> (<em>finales de los 80</em>) – Subgénero de la EDM, la música house se caracteriza por su ritmo 4/4, líneas de bajo sintetizadas y bucles repetitivos.</li>
<li><strong>Techno</strong> (<em>finales de los 80</em>) – Otro subgénero de la EDM, el techno es conocido por sus ritmos repetitivos, sonidos sintetizados y un enfoque en el ritmo por encima de la melodía.</li>
<li><strong>Grunge</strong> (<em>finales de los 80-1990</em>) – Subgénero del rock alternativo, el grunge se caracteriza por su sonido de guitarra pesado y distorsionado y sus letras introspectivas, a menudo angustiosas.</li>
<li><strong>Britpop</strong> (<em>años 90</em>) – Movimiento británico de rock alternativo caracterizado por melodías pegadizas, sonido
guitarrero y un sentimiento de orgullo nacional, a menudo visto como una reacción al grunge estadounidense.</li>
<li><strong>Trance</strong> (<em>década de 1990</em>) – Subgénero de la EDM. El trance se caracteriza por frases melódicas repetitivas, ritmos hipnóticos y oleadas que conducen a clímax enérgicos.</li>
<li><strong>Indie Rock</strong> (<em>década de 1990-actualidad</em>) – El indie rock, un género diverso que se originó como reacción al rock dominante, a menudo presenta sonidos y estructuras poco convencionales, así como una fuerte ética del bricolaje.</li>
<li><strong>Pop Punk</strong> (<em>década de 1990-2000</em>) – Fusión de punk rock y música pop, el pop punk se caracteriza por melodías alegres, ganchos pegadizos y actuaciones enérgicas.</li>
<li>**Emo **(<em>década de 1990-2000</em>) – Subgénero del punk y el indie rock, conocido por sus letras introspectivas y emotivas, sus composiciones confesionales y sus voces expresivas.</li>
<li><strong>Post-Rock</strong> (<em>1990-presente</em>) – Género experimental caracterizado por el uso de instrumentos de rock para crear paisajes sonoros atmosféricos y cinematográficos, a menudo con letras mínimas y estructuras de canción poco convencionales.</li>
<li><strong>K-pop</strong> (<em>1990-presente</em>) – Originario de Corea del Sur, el K-pop se caracteriza por melodías pegadizas, coreografías intrincadas y vídeos musicales visualmente impactantes, a menudo interpretados por grupos de ídolos muy pulidos.</li>
<li><strong>Reggaetón</strong> (<em>años 90-actualidad</em>) – El reggaetón es una fusión de reggae, hip hop y estilos musicales latinoamericanos. Se caracteriza por su ritmo dembow y sus letras a menudo explícitas.</li>
<li><strong>Dubstep</strong> (<em>principios de la década de 2000)</em> – Surgió del garage británico y del drum and bass. El dubstep se caracteriza por fuertes líneas de bajo, ritmos sincopados y poco uso de la voz.</li>
</ul>  </div>
<h2 id="subgéneros-musicales">Subgéneros musicales</h2>
<p>A continuación se enumeran algunos subgéneros destacados dentro de los géneros principales mencionados anteriormente:</p>
<h3 id="rock">Rock</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Rock alternativo</li>
<li>Arte Rock</li>
<li>Garage Rock</li>
<li>Glam Rock</li>
<li>Grunge</li>
<li>Hard Rock</li>
<li>Indie Rock</li>
<li>Math Rock</li>
<li>Post-Rock</li>
<li>Rock Progresivo</li>
<li>Rock Psicodélico</li>
<li>Punk Rock</li>
<li>Rock Sureño</li>
<li>Stoner Rock</li>
<li>Shoegaze</li>
<li>Post-Punk</li>
<li>No Wave</li>
</ol>  </div>
<h3 id="metal">Metal</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Black Metal</li>
<li>Death Metal</li>
<li>Doom Metal</li>
<li>Folk Metal</li>
<li>Glam Metal</li>
<li>Metal gótico</li>
<li>Metal industrial</li>
<li>Power Metal</li>
<li>Metal progresivo</li>
<li>Sludge Metal</li>
<li>Metal sinfónico</li>
<li>Thrash Metal</li>
</ol>  </div>
<h3 id="música-electrónica-de-baile-edm">Música electrónica de baile (EDM)</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Ambiente</li>
<li>Breakbeat</li>
<li>Drum and Bass</li>
<li>Dubstep</li>
<li>Electro</li>
<li>Glitch</li>
<li>Hardcore</li>
<li>House</li>
<li>IDM (Intelligent Dance Music)</li>
<li>Jungle</li>
<li>Techno</li>
<li>Trance</li>
<li>Trap</li>
<li>Trip Hop</li>
<li>Vaporwave</li>
<li>Chillwave</li>
<li>Darkwave</li>
<li>EBM (Electronic Body Music)</li>
</ol>  </div>
<h3 id="hip-hop">Hip Hop</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Hip Hop alternativo</li>
<li>Boom Bap</li>
<li>Hip Hop consciente</li>
<li>Crunk</li>
<li>Dirty South</li>
<li>Hip Hop de la Costa Este</li>
<li>Gangsta Rap</li>
<li>G-Funk</li>
<li>Grime</li>
<li>Hip Hop Hardcore</li>
<li>Horrorcore</li>
<li>Jazz Rap</li>
<li>Hip Hop Old School</li>
<li>Trap</li>
<li>Hip Hop Costa Oeste</li>
</ol>  </div>
<h3 id="country">Country</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Alt-Country</li>
<li>Americana</li>
<li>Bluegrass</li>
<li>Country clásico<br>
Country pop</li>
<li>Country Rock</li>
<li>Honky Tonk</li>
<li>Outlaw Country</li>
<li>Country tradicional</li>
</ol>  </div>
<h3 id="jazz">Jazz</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Jazz ácido</li>
<li>Jazz de vanguardia</li>
<li>Bebop</li>
<li>Jazz fresco</li>
<li>Dixieland</li>
<li>Jazz libre</li>
<li>Fusión</li>
<li>Jazz gitano</li>
<li>Jazz latino</li>
<li>Jazz modal</li>
<li>Jazz suave</li>
<li>Swing 13</li>
<li>Jazz vocal</li>
</ol>  </div>
<h3 id="blues">Blues</h3>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ol>
<li>Acoustic Blues</li>
<li>Chicago Blues</li>
<li>Country Blues</li>
<li>Delta Blues</li>
<li>Electric Blues</li>
<li>Gospel Blues</li>
<li>Jump Blues</li>
<li>Piedmont Blues</li>
<li>Texas Blues</li>
</ol>  </div>
<h2 id="estilos-musicales-regionales">Estilos musicales regionales</h2>
<p>Con los estilos regionales ocurre lo mismo que con los estilos y subgéneros generales: es imposible identificar cada uno de ellos con exactitud, ya que siempre se proponen otros nuevos.</p>
<p>No obstante, a continuación se enumeran algunos estilos regionales populares e influyentes en distintas partes del mundo:</p>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="norte-américa">Norte América</h3><ul>
<li>Folclore de los Apalaches</li>
<li>Cajún</li>
<li>Criollo</li>
<li>Blues del Delta</li>
<li>Música nativa americana</li>
<li>Jazz de Nueva Orleans</li>
<li>Tejano</li>
<li>Tex-Mex</li>
<li>Zydeco</li>
</ul>  </div>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="sudamérica">Sudamérica</h3><ul>
<li>Tango argentino</li>
<li>Bossa Nova (Brasil)</li>
<li>Candombe (Uruguay)</li>
<li>Chacarera (Argentina)</li>
<li>Choro (Brasil)</li>
<li>Cumbia (Colombia)</li>
<li>Forró (Brasil)</li>
<li>Merengue (República Dominicana)</li>
<li>Samba (Brasil)</li>
<li>Salsa (Cuba, Puerto Rico)</li>
<li>Vallenato (Colombia)</li>
</ul>  </div>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="europa">Europa</h3><ul>
<li>Balkan Brass Band (Sudeste de Europa)</li>
<li>Música celta (Irlanda, Escocia, Gales, Bretaña)</li>
<li>Fado (Portugal)</li>
<li>Flamenco (España)</li>
<li>Folclore griego (Grecia)</li>
<li>Klezmer (Europa del Este)</li>
<li>Folclore nórdico (Escandinavia)</li>
<li>Polka (Europa Central)</li>
<li>Música romaní (Europa del Este)</li>
<li>Tarantela (Italia)</li>
<li>Canto de garganta tuvan (República de Tuva, Rusia)</li>
<li>Mugham (Azerbaiyán)</li>
</ul>  </div>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="áfrica">África</h3><ul>
<li>Afrobeat (África Occidental)</li>
<li>Benga (Kenia)</li>
<li>Gnawa (Marruecos)</li>
<li>Highlife (África Occidental)</li>
<li>Juju (Nigeria)</li>
<li>Kwaito (Sudáfrica)</li>
<li>Mbalax (Senegal)</li>
<li>Rai (Argelia)</li>
<li>Soukous (Congo)</li>
<li>Taarab (África Oriental)</li>
<li>Makossa (Camerún)</li>
</ul>  </div>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="asia">Asia</h3><ul>
<li>Bhangra (India)</li>
<li>Ópera china (China)</li>
<li>Gamelan (Indonesia)</li>
<li>Gagaku (Japón)</li>
<li>K-Pop (Corea del Sur)</li>
<li>Qawwali (Asia Meridional)</li>
<li>Tala (India)</li>
<li>Música clásica tailandesa (Tailandia)</li>
<li>Música tradicional japonesa (Japón)</li>
<li>Música tradicional persa (Irán)</li>
<li>Dangdut (Indonesia)</li>
<li>Música sufí (Oriente Próximo, Norte de África, Sur de Asia)</li>
<li>Enka japonés (Japón)</li>
<li>J-Pop (Japón)</li>
<li>Cantopop (Hong Kong)</li>
<li>Mandopop (China continental, Taiwán)</li>
</ul>  </div>
<div class="rounded-md p-4 !my-5 bg-success-50 text-success-700 dark:bg-success-950 dark:text-success-300">  <h3 id="oceanía">Oceanía</h3><ul>
<li>Música aborigen (Australia)</li>
<li>Música hawaiana (Hawai)</li>
<li>Música maorí (Nueva Zelanda)</li>
<li>Música de las islas del Pacífico (Polinesia, Micronesia, Melanesia)</li>
</ul>  </div>
<p><img src="/_vercel/image?url=_astro%2Flinkedin-musica-80x80.PfETbYec.png&amp;w=640&amp;q=100" alt="linkedin musica" class="float-left mb-4 mr-4" width="80" height="80" loading="lazy" decoding="async"></p>
<p>Como se puede observar, existen numerosos estilos musicales que han ido cambiando a lo largo de los años y diferenciándose por regiones, dependiendo de la situación social del momento y de los intereses de sus gentes.</p>
<p>Así, podemos encontrar una rica y gran variedad de estilos musicales, para poder elegir y disfrutar de los que más nos gusten.</p>
<h2 id="referencias">Referencias</h2>
<div class="rounded-md p-4 !my-5 bg-info-50 text-info-700 dark:bg-info-950 dark:text-info-300">  <ul>
<li><a href="https://www.significados.com/generos-musicales/" target="_blank" rel="noopener"> Los 25 géneros musicales más populares.  </a>Enciclopedia Significados</li>
</ul>  </div>  </div>
```
